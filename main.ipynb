```python
import uproot
import os 
import pandas as pd
import awkward as ak
import glob 
```

```python
mc_path = 'dataset/4lep 2/MC/'
mc_files = glob.glob(os.path.join(mc_path, '*.root'))
mc_df_list = []
```

```python
print("Processing Monte Carlos")
for filepath in mc_files:
    filename = os.path.basename(filepath)
    print(f" -> Loading{filename}")


    with uproot.open(filepath) as file:
        df = file['mini'].arrays(library='pd')
        

        if 'signal' in filename:
            df['label']= 1
        else:
            df['label']= 0 

        mc_df_list.append(df)
```

```python
mc_df = pd.concat(mc_df_list, ignore_index = True)
```

```python
print("\n MC data loaded successfully!")
print(f"Total MC events : {len(mc_df)}")
print("Label distribution:")
print(mc_df['label'].value_counts())
```
